```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
# load the data set
tips=sn.load_dataset("tips")
tips.head()
```

```python
tips['tip'].mean()
```

```python
sn.stripplot(x=tips['tip'],color="green")
plt.show()
```

```python
sn.stripplot(x=tips["total_bill"])
plt.show()
```

```python
sn.stripplot(data=tips)
plt.show()
```

```python
# for two varibales
sn.stripplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
# for more dispersed value
sn.stripplot(x='day',y='total_bill',data=tips,jitter=False)
plt.show()
```

```python
# we can controll all jitter distribution 
sn.stripplot(x='day',y='total_bill',data=tips,jitter=0.2)
```

```python
# for horizontal graph we can change by two method
# change x,y
# change orientation
sn.stripplot(x='total_bill',y="day",data=tips)
```

```python
sn.stripplot(x='total_bill',y="day",data=tips,hue='sex')
```

```python
#draw some outline over there
sn.stripplot(x='total_bill',y='day',data=tips,linewidth=10)
```

```python
sn.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sn.stripplot(x='day',y="total_bill",data=tips,hue='smoker')
plt.show()
```

```python
# to deal with warnings 
import warnings 
warnings.filterwarnings("ignore")
# for separate the data of sex
sn.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
plt.show()
```

```python
# for smoker
sn.stripplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
plt.show()
```

```python
sn.stripplot(x='day',y='total_bill',data=tips,hue='smoker',split=True,order=['Fri',"Sun"])
plt.show()
```

```python
sn.stripplot(x='sex',y='tip',data=tips,order=["Female","Male"],palette='spring')
plt.show()
```

```python
#for different markers 
sn.stripplot(x='sex',y='tip',data=tips,order=["Female","Male"],palette='spring',marker="D")
plt.show()
```

```python
sn.stripplot(x='day',y='total_bill',data=tips,marker="D",size=13,hue='sex',split=True)
```

```python
sn.stripplot(x='day',y='total_bill',data=tips,marker="D",size=13,hue='sex',edgecolor='gray',alpha=0.3,palette="husl")
plt.show()
```

```python
# for combination of stripplot and boxplot
sn.stripplot(x='tip',y='day',data=tips)
sn.boxplot(x='tip',y='day',data=tips)
```

```python
sn.stripplot(x='tip',y='day',data=tips)
sn.violinplot(x='tip',y='day',data=tips,color="0.9")
```
